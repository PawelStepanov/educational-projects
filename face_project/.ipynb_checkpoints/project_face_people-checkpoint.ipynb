# Разработка системы компьютерного зрения для обработки фотографий

Сетевой супермаркет «Хлеб-Соль» внедряет систему компьютерного зрения для обработки фотографий покупателей. Фотофиксация в прикассовой зоне поможет определять возраст клиентов, чтобы:
- Анализировать покупки и предлагать товары, которые могут заинтересовать покупателей этой возрастной группы;
- Контролировать добросовестность кассиров при продаже алкоголя.

**Цель** - создать систему компьютерного зрения для обработки фотографий покупателей.

**Задача:**
- Построить модель, которая по фотографии определит приблизительный возраст человека. 

В нашем распоряжении набор фотографий людей с указанием возраста. Данные взяты с сайта [ChaLearn Looking at People](https://chalearnlap.cvc.uab.es/dataset/26/description/). CSV-файл `labels.csv` с двумя колонками: `file_name` и `real_age`. 

**План исследования**

1. Изучение данных 
2. Обучение модели
3. Вывод

## Изучение данных

### Импорт библиотек

````python
import os
import warnings

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro
from keras.layers import Dense, GlobalAveragePooling2D
from keras.applications.resnet import ResNet50
from keras.models import Sequential
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
````

````python
%matplotlib inline
warnings.simplefilter('ignore', UserWarning)
warnings.simplefilter('ignore', FutureWarning)
````

### Константы

````python
SEED = 12345
ALPHA = 0.05
PTH = '/datasets/faces/labels.csv'
PTH1 = '/datasets/faces/final_files'
````

### Загрузка данных

Сохраним файл `labels.csv` в переменную `labels`

````python
if os.path.exists(PTH):
    labels = pd.read_csv(PTH)
else:
    print('Файл не найден')
````

Посмотрим на первые 5 строк

````python
labels.head()  # Смотрим на первые 5 строк
````

Посмотрим какого размера у нас выборка.

````python
labels.shape  # Смотрим на размер файла
````

Всего у нас 7591 фотография. Для загрузки изображений воспользуемся динамической загрузкой данных.

````python
datagen = ImageDataGenerator(validation_split=0.25, rescale=1./255)
````

````python
train_datagen_flow = datagen.flow_from_dataframe(
    dataframe=labels,
    directory=PTH1,
    x_col="file_name",
    y_col="real_age",
    target_size=(150, 150),
    batch_size=16,
    class_mode='raw',
    subset='training',
    seed=SEED
)
````

Теперь изучим возраст и построим гистограмму распределения и график с усами. А также посмотрим на фотографии в датасете.

### Исследовательский анализ данных

Посмотрим диаграмму размаха для визуального представления групп числовых данных через квартили и гистограмму распределения.

````python
plt.figure(figsize=(15, 10))

plt.subplot(2, 1, 1)
sns.boxplot(labels['real_age'], width=.1)
plt.title('Диаграмма размаха по возрасту')

plt.subplot(2, 1, 2)
sns.histplot(labels['real_age'], bins=100)
plt.title('Гистограмма распределения по возрасту')
plt.show()
````

В целом данные выглядят нормально, возраст людей не превышает 100, но есть значения близкие к этому возрасту, такое в принципе возможно, но очень редко. Теперь посмотрим на распределение данных. Большинство людей в возрасте от 20 до 40 лет.

Данные распределенны ненормально, но чтобы быть увереным точно проведем статистический тест на нормальность Шапиро-Уилка. Необходимо выдвинуть две гипотезы:
- **H0** - данные получены из нормального расрпеделения 
- **H1** - Данные не получены из нормального распределения

Предположим, что верна **H0**, тогда

````python
stats, p_value = shapiro(labels['real_age'])

if p_value > ALPHA:
    print('Не отвергаем H0. Данные получены из нормального распределения')
else:
    print('Отвергаем Н0 и принимаем Н1. Данные не получены из нормального распределения')
````

Тест также доказал, что данные распределены ненормально

Теперь посмотрим на фотографии в датасете.

````python
# Получаем объект картинка-метка
features, target = next(train_datagen_flow)
````

````python
# Выводим 16 изображений
fig = plt.figure(figsize=(10, 10))
for i in range(16):
    fig.add_subplot(4, 4, i+1)
    plt.imshow(features[i])
    plt.xlabel(f'Возраст: {target[i]}')
    plt.xticks([])
    plt.yticks([])
    plt.tight_layout()
````

### Вывод

По результатам исследования можно сказать, что у нас всего 7591 фотографий людей разных возрастов, при это людей возраста > 80 немного и это может повлиять на результаты моделей.

Фотографии в датасете разного качества, какие-то хорошие, какие-то размытые, на каких-то легко виден человек, но на некоторых человека плохо видно (4 фотография). С учетом того, что у нас немного фотографий стоит проделать аугментацию. 

## Обучение модели 

Можем приступать к обучению модели. В качестве модели будем использовать `ResNet50` с инициализацией весов (`ImageNet`) и без верхушки сети, чтобы адаптировать сеть к нашей задаче. Так как перед нами задача регрессии, то на выходе должен быть один нейрон с функцией активацией `linear`. В качестве функции потерь будем использовать `MAE` желательно, чтобы она была меньше 8.

Так как для задач CV необходимо множество вычислительных ресурсов, вычисления проводились на серверах Яндекса, и для этого необходимо было написать 4 функции.
- загрузки обучающей выборки `load_train()`,
- загрузка тестовой выборки `load_test()`
- создания модели `create_model()`,
- запуска модели `train_model()`.

Результат вывода модели приложен после функций

```python
def load_train(path):
    '''
    Функция для загрузки обучающей выборки, передается 
    путь папки, в которой лежит csv файл и папка с изображениями
    на выходе результат работы загрузчика
    '''
    
    # Сохраняем путь к csv файлу и читаем его
    labels_path = f"{path}/labels.csv"
    df = pd.read_csv(labels_path)

    # Задействуем загрузчик данных и делаем аугментацию
    datagen = ImageDataGenerator(validation_split=0.25, rescale=1./255,
                                 rotation_range=15, width_shift_range=0.1,
                                 height_shift_range=0.1)
    
    # Сохраняем путь к папке с изображениями
    face_path = f"{path}/final_files"
    train_datagen_flow = datagen.flow_from_dataframe(
        dataframe=df,
        directory=face_path,
        x_col="file_name",
        y_col="real_age",
        target_size=(224, 224),
        batch_size=16,
        class_mode='raw',
        subset='training',
        seed=12345)

    return train_datagen_flow
```

```python
def load_test(path):
    '''
    Функция для загрузки тестовой выборки, передается
    путь папки, в которой лежит csv файл и папка с изображениями
    на выходе результат работы загрузчика
    '''
    
    # Сохраняем путь к csv файлу и читаем его
    labels_path = f"{path}/labels.csv"
    df = pd.read_csv(labels_path)
    
    # Задействуем загрузчик данных
    datagen = ImageDataGenerator(validation_split=0.25, rescale=1./255)
    
    # Сохраняем путь к папке с изображениями
    face_path = f"{path}/final_files"
    valid_datagen_flow = datagen.flow_from_dataframe(
        dataframe=df,
        directory=face_path,
        x_col='file_name',
        y_col='real_age',
        target_size=(224, 224),
        batch_size=16,
        class_mode='raw',
        subset='validation',
        seed=12345)

    return valid_datagen_flow
```

```python
def create_model(input_shape):
    '''
    Функция для создания модели, передается 
    размер входа, на выходе получаем модель
    '''

    # Основная часть модели
    backbone = ResNet50(input_shape=input_shape,
                        weights='imagenet',
                        include_top=False)

    model = Sequential()
    
    # Инициализируем оптимизатор
    optimizer = Adam(lr=0.0001)
    
    # Добавляем базу, глобальный пулинг и слой для регрессии
    model.add(backbone)
    model.add(GlobalAveragePooling2D())
    model.add(Dense(1, activation='linear'))
    
    # Компилируем модель
    model.compile(optimizer=optimizer, loss='mse',
                  metrics=['mae'])
    return model
```

```python
def train_model(model, train_data, test_data, batch_size=None, epochs=10,
               steps_per_epoch=None, validation_steps=None):
    '''
    Функция для обучения модели. Принимает 7 аргументов
    на выходе обученая модель.
    '''
    
    model.fit(train_data, validation_data=test_data,
              batch_size=batch_size, epochs=epochs,
              steps_per_epoch=steps_per_epoch,
              validation_steps=validation_steps,
              verbose=2)

    return model 
```
